# How many references does a given proof require?
    - idea: related to short-term memory limit of 7 chuncks of info.

```python
import json
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from net_metrics import net_metrics
```

```python
#load proofs network
with open("data/ProofWiki-proofs-only.json", "r") as d:
    proofs_dict = json.load(d)

#proofs_dict: proof title --> list of theorems used in proof
```

```python
#build dataframe with references
proofs_df = pd.DataFrame(proofs_dict.items())
proofs_df.columns = ['article', 'references']

proofs_df['number of references'] = proofs_df['references'].apply(lambda x: len(x))
proofs_df
```

```python
proofs_df['number of references'].describe()
```

## Which Theorems require the most references?


```python
proofs_df = proofs_df.sort(columns='number of references', ascending=False)
proofs_df
```

```python
plt.figure(figsize=(8,6))

sns.violinplot(y='number of references', data=proofs_df)
```
