```python
#create a network using only the first theorem used in a proof

proofs_dict_single = {}

for proof, theorem in proofs_dict.iteritems():
    proofs_dict_single[proof] = theorem[0]
    
```

## Accumulation of First References 
**Traversal Visits**

measured as the accumulated number of visits after traversing the network:
 1. begin at every proof
 2. first reference receives +1 visit
 3. continue to next reference, +1 visit 
     - stop at a repeated proof or dead end

```python
traversal_visits = net_metrics.compute_traversal_visits(proofs_dict_single)
```

```python
traversal_visits_df = pd.DataFrame(traversal_visits.items())
traversal_visits_df
```

```python
traversal_visits_df.columns = ['proof', 'visits']

traversal_visits_df = traversal_visits_df.sort(columns='visits', ascending=False)
```
